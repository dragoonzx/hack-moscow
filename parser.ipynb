```python
import requests as req

def get_page(s,id,page=1):
    return req.get(s,params={"person": id,
                             "page":page}).json()

def get_page(s, page=1):
    return req.get(s,params={"page":page}).json()

def parser(element):
    d = {"name":element['main']['person']['name'],
            "incomes":element['incomes'][0]['size'],
            "squares":[]}
    for i in element['real_estates']:
        d['squares'].append(i['square'])
    return d


s = []
url = "https://declarator.org/api/v1/search/sections/?region"
id_region = 63

params = {"region": id_region,
          "year": 2016}
```

```python
r = req.get(url,params={})
if r.status_code == 200:
    count = r.json()['count']
    for i in range(1,(count+19)//10):
        p = get_page(url, page=i)
        for j in p['results']:
            s.append(parser(j))
        if i == 50: 
            break
        print (i)
    print(len(s))
    print(s[0])
else:
    print(r.status_code)
```

```python
import pandas as pd
import numpy as np
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
output_notebook()

def make_plot(title, hist, edges, x):
    p = figure(title=title, tools='', background_fill_color="#fafafa")
    p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
           fill_color="navy", line_color="white", alpha=0.5)

    p.y_range.start = 0
    p.legend.location = "center_right"
    p.legend.background_fill_color = "#fefefe"
    p.xaxis.axis_label = 'Зарплата'
    p.yaxis.axis_label = 'Кол-во человек'
    p.grid.grid_line_color="white"
    return p

df = pd.DataFrame(s)

hist, edges = np.histogram(df.incomes, density=False, bins=100)
print(df.describe())
x = np.linspace(df.incomes.min(), df.incomes.max(),100)
show(make_plot("sdfsdf",hist, edges,x))
```
